## Data Modeling
- Mathematical/statistical model used:
    - I used Pearson Corelation Coefficients that I ended up coding in myself instead of using the package because I has having problems debugging code and mine was more straightforward and easier to diagnose problems in and uses the Pearson Corelation Coefficient formula to find the r values
        - I used Pearson corelation coefficients because I wanted to be able to find if there was any corelation in the data and Pearson corelation coefficients is a terrific way to do that
- I found the corelation in race, gender, and armed status, vs cause of death 
    - I broke up armed status so that it was whether the victim was armed 
        - It had originally been a variety of things such as unknown, but I had done corelations with that and it had been a pretty low r 
    - I also broke up cause of death to whether they were shot in
        - I had originally broken it into almost all the groups that the data gave me which was a lot, but again the corelation was low so I grouped it 



```python
def sumOfData(data):
    i=0
    total=0
    n=len(data)
    while(i<n):
        total = total +data[i]
        
        i=i+1
    return total


def sumOfSqr(data):
    i=0
    total=0
    n=len(data)
    while(i<n):
        total = total + (data[i]*data[i])
        
        i=i+1
    return total

def sumOfProductOfPairs(data1, data2):
    i=0
    total=0
    n=len(data1)
    if(n!=len(data2)):
        raise ValueError('Data is of different lengths aaah')
                         
    while(i<n):
        total = total + (data1[i]*data2[i])
        i=i+1
    return total

def pearson(x,y):
    n=len(x)
    if(n!=len(y)):
        raise ValueError('x,y is of different lengths aaah')
    
    sumOfx=sumOfData(x) 
    sumOfy=sumOfData(y)
                         
    a=(n * sumOfProductOfPairs(x,y)) - (sumOfx * sumOfy)
    b=((n * sumOfSqr(x)) - (sumOfx * sumOfx))
    c=((n * sumOfSqr(y)) - (sumOfy * sumOfy))                  
    
    r=a/(math.sqrt(b*c))
    return r

r = pearson(kbp['Victims race'], kbp['Cause of death'])
print("race vs cause of death rval: ",r)
r = np.corrcoef(kbp['Victims race'], kbp['Cause of death'], True)[0, 1]
print("race vs cause of death rval2: ",r)


r = pearson(kbp['Victims gender'], kbp['Cause of death'])
print("gender vs cause of death rval: ",r)
r = np.corrcoef(kbp['Victims gender'], kbp['Cause of death'], True)[0, 1]
print("gender vs cause of death rval2: ",r)


r = pearson(kbp['Unarmed'], kbp['Cause of death'])
print("armed vs cause of death rval: ",r)
r = np.corrcoef(kbp['Unarmed'], kbp['Cause of death'], True)[0, 1]
print("armed vs cause of death rval2: ",r)
```
